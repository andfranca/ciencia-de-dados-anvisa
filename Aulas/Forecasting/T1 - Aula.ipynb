```python
#!pip install yahooquery
from yahooquery import Ticker
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
ticker = ['PETR4.SA','ITUB4.SA']
bd = Ticker(ticker)
bd = bd.history(start='2000-01-01',end='2023-07-11')
petr = bd.loc['PETR4.SA',:]
```

```python
petr[['close']].plot()
petr[['open']].plot()
```

```python
petr.head()
```

```python
config = {
    'x': petr.index,
    'open': petr.open,
    'close': petr.close,
    'high': petr.high,
    'low': petr.low,
    'type':'candlestick',
}

data = [config]
layout = go.Layout()
fig = go.Figure(data=data,
                layout=layout)
fig.show()
```

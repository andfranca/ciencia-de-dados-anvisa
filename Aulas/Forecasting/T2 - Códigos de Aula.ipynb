```python
from prophet import Prophet
from yahooquery import Ticker
import matplotlib.pyplot as plt
```

```python
inicio = '2016-01-01'
fim = '2023-07-19'

acao = ['PETR4.SA','ITUB3.SA']
bd = Ticker(acao)
bd = bd.history(start=inicio,end=fim)
petr = bd.loc['PETR4.SA',:]
```

```python
petr2 = petr[['close']].reset_index()
petr2.columns = ['ds','y']
petr2['ds'] = pd.to_datetime(petr2['ds'])
```

```python
petr2.head()
```

```python
plt.plot(petr2['ds'],petr2['y'])
plt.show()
```

```python
modelo = Prophet()
modelo_p = modelo.fit(petr2)
```

```python
futuro = modelo_p.make_future_dataframe(periods=12,freq='M')
y_pred = modelo_p.predict(futuro)
plot1 = modelo_p.plot(y_pred)
```

#### Treino e Teste

```python
train = petr2[petr2['ds']<='2022-12-31']
test = petr2[petr2['ds']>'2022-12-31']
```

```python
modelo = Prophet()
modelo_p = modelo.fit(train)
y_pred = modelo_p.predict(test)
```

```python
y_obs = test['y']
y_pred1 = y_pred['yhat']
```

```python
from sklearn import metrics
```

```python
mse = metrics.mean_squared_error(y_obs,y_pred['yhat'], squared=True)
rmse = metrics.mean_squared_error(y_obs,y_pred['yhat'], squared=False)
mae = metrics.mean_absolute_error(y_obs,y_pred['yhat'])
r2 = metrics.r2_score(y_obs,y_pred['yhat'])

print(f'mse: {mse}')
print(f'rmse: {rmse}')
print(f'mae: {mae}')
print(f'R2: {r2}')
```

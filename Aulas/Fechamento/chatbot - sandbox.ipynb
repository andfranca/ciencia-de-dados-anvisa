```python
import spacy
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
from nltk.stem import RSLPStemmer
import time
time.clock = time.time
```

```python
# Carregar o modelo em português do SpaCy
nlp = spacy.load("pt_core_news_sm")

# Criar o chatbot
bot = ChatBot('MeuChatBot')

# criar a função de pré-processamento com o SpaCy e o RSLP
def spacy_rslp_preprocessor(statement):
    text = statement.text
    doc = nlp(text)
    # Juntar as palavras lematizadas e aplicar o RSLP
    rslp = RSLPStemmer()
    statement.text = " ".join(rslp.stem(token.lemma_) for token in doc)
    return statement

# Adicionar a função de pré-processamento personalizada ao chatbot
bot.preprocessors.append(spacy_rslp_preprocessor)

# Criar e treinar o treinador do chatbot com o corpus em português
trainer = ChatterBotCorpusTrainer(bot, show_training_progress=False)
trainer.train('chatterbot.corpus.portuguese')

# Interagir com o chatbot
while True:
    user_input = input("Você: ")
    if user_input.lower() == 'sair':
        break
    response = bot.get_response(user_input)
    print("ChatBot: " + str(response))
```

```python
#Tkinter e chatgpt
import tkinter as tk
from tkinter import messagebox
import openai

# Defina sua chave de API aqui
api_key = "CHAVE_API"

# Configuração da API do OpenAI
openai.api_key = api_key

def chat_with_bot(prompt):
    try:
        response = openai.Completion.create(
            engine="text-davinci-003",  # Você também pode tentar "text-davinci-002" para versões mais recentes do GPT-3
            prompt=prompt,
            max_tokens=150  # Define o limite de tokens para a resposta do chatbot
        )

        return response.choices[0].text.strip()
    except Exception as e:
        return str(e)

def send_message():
    user_input = entry.get().strip()
    if not user_input:
        return

    prompt = f"Você disse: {user_input}\nChatbot disse:"
    response = chat_with_bot(prompt)
    chat_history.append(f"Você: {user_input}")
    chat_history.append(f"Chatbot: {response}")
    update_chat_history()

def update_chat_history():
    chat_history_text.set("\n".join(chat_history))
    entry.delete(0, tk.END)

if __name__ == "__main__":
    chat_history = []
    root = tk.Tk()
    root.title("Chatbot usando Tkinter")

    chat_history_text = tk.StringVar()

    chat_history_label = tk.Label(root, textvariable=chat_history_text, wraplength=400, justify="left")
    chat_history_label.pack(padx=10, pady=10)

    entry = tk.Entry(root, width=50)
    entry.pack(padx=10, pady=5)

    send_button = tk.Button(root, text="Enviar", command=send_message)
    send_button.pack(pady=5)

    root.geometry("400x300")
    root.mainloop()
```

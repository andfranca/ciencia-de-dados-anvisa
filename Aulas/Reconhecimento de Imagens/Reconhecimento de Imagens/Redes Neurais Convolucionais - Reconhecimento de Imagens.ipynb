## Reconhecimento de Imagem

O reconhecimento de imagem, também conhecido como visão computacional, é uma das áreas com mais destaque na inteligência artificial e envolve a capacidade de um sistema compreender e interpretar padrões visuais em imagens ou vídeos. O objetivo do reconhecimento de imagem é permitir que os computadores "vejam" o mundo de uma forma semelhante aos seres humanos, identificando e compreendendo objetos, padrões e características visuais.

<img src="https://drive.google.com/uc?id=1sVmnKOI7LHV-v1Z29KgYdWWOt1VY3HpX" width="400" align="center"/>

O processo de reconhecimento de imagem geralmente envolve a utilização de algoritmos e técnicas de aprendizado de máquina, como redes neurais convolucionais (CNNs), para analisar e extrair informações relevantes das imagens. Esses algoritmos são treinados em grandes conjuntos de dados contendo imagens rotuladas, permitindo que eles aprendam a reconhecer padrões e características específicas.

O reconhecimento de imagem tem diversas aplicações práticas, incluindo:

- **Classificação de objetos:** identificar e categorizar objetos em imagens, como reconhecimento de rostos, detecção de veículos, identificação de animais, entre outros.
- **Detecção e localização de objetos:** identificar a presença e localização de objetos específicos em imagens ou vídeos.
- **Segmentação de imagem:** separar uma imagem em regiões ou objetos diferentes com base em suas características visuais.
- **Reconhecimento de texto:** extrair texto de imagens, como leitura de placas de veículos, reconhecimento óptico de caracteres (OCR), entre outros.
- **Análise de sentimentos:** identificar expressões faciais e emoções em imagens de rostos humanos.
- **Realidade aumentada:** sobrepor informações virtuais em tempo real em cima de imagens do mundo real.


### Classificação de Imagens

A classificação de imagem é uma tarefa da área de visão computacional que envolve atribuir rótulos ou categorias a imagens com base em suas características. O objetivo é ensinar um sistema a reconhecer e classificar automaticamente as imagens em diferentes classes ou categorias pré-definidas.

Na classificação de imagem, um algoritmo é treinado em um conjunto de dados de treinamento que contém imagens já rotuladas. Cada imagem é associada a uma categoria ou classe específica, como "cachorro", "gato", "carro", "pessoa", e assim por diante. O algoritmo aprende a identificar os padrões e características visuais nas imagens de treinamento que são discriminantes para cada classe.

Após o treinamento, o algoritmo pode ser aplicado a novas imagens para fazer a classificação. Com base nas características extraídas das imagens de teste, o algoritmo atribui um rótulo ou categoria à imagem, indicando a classe à qual ela pertence.

#### Tipos de classificação

Dependendo do problema, existem diferentes tipos de metodologias de classificação de imagens que podem ser utilizadas, como por exemplo: 

‍**Binário:** a classificação binária usa uma lógica de 0 ou 1 para rotular imagens e classifica pontos de dados desconhecidos em duas categorias. Muito utilizada para categorizar tumores benignos ou malignos, analisar se um medicamento gera efeito adverso ou não, e muitas outras tarefas que exigem respostas sim/não.‍

**Multiclasse:** enquanto a classificação binária diferencia apenas duas classes, o multiclasse, categoriza itens em três ou mais classes. É muito útil para prever tipos de reações ou diagnóstico médico.

**Multilabel:** Enquanto na multiclasse cada imagem é atribuída a exatamente uma classe, a classificação multilabel permite que o item seja atribuído a vários rótulos. Por exemplo, pode ser necessário classificar as cores da imagem e existem várias cores. Uma foto do arco-íris terá vermelho, amarelo, azul e outras cores. Como resultado, uma imagem terá várias cores como rótulos.‍

**Hierárquico:** A classificação hierárquica é a tarefa de organizar as classes em uma estrutura hierárquica com base em suas semelhanças, onde uma classe de nível superior representa categorias mais amplas e uma classe de nível inferior é mais concreta e específica.

### Classificação de Imagens vs Detecção de objeto

Ao entrar no mundo da visão computacional é comum nos depararmos com os termos classificação de imagem e detecção de objetos, porém, essas duas áreas possuem funções diferentes. 

A classificação de imagem envolve atribuir uma única categoria ou classe a uma imagem completa. O objetivo é identificar a categoria à qual a imagem pertence. Por exemplo, dada uma imagem de um gato, o objetivo da classificação de imagem seria atribuir a categoria "gato" a essa imagem. A classificação de imagem assume que a imagem contém apenas um objeto e o rótulo atribuído é a classe predominante na imagem.

Já a detecção de objetos tem como objetivo identificar e localizar a presença de objetos específicos em uma imagem. Diferentemente da classificação de imagem, a detecção de objetos não se limita a uma única categoria ou classe, e ela pode identificar múltiplos objetos de diferentes categorias em uma única imagem. A detecção de objetos fornece informações sobre a localização dos objetos detectados, geralmente por meio de retângulos delimitadores (bounding boxes) que cercam cada objeto encontrado. Além disso, a detecção de objetos também é capaz de lidar com cenários em que há sobreposição ou interseção de objetos na imagem.

### Redes Neurais Convolucionais

Uma rede neural convolucional (_CNN - Convolutional Neural Network_) é um tipo de arquitetura de rede neural projetada para processar dados com estrutura de grade, como imagens. Ela é amplamente utilizada na área de visão computacional e reconhecimento de padrões.

As CNNs são inspiradas pelo funcionamento do córtex visual do cérebro humano, onde os neurônios individuais respondem a regiões restritas do campo visual, chamadas de campos receptivos. Da mesma forma, as CNNs têm camadas especializadas, como camadas de convolução, que ajudam a extrair e aprender características relevantes de uma imagem.


Na CNN, cada imagem é representada como uma matriz de valores de pixel. Uma imagem digital é composta por pixels, que são os elementos mínimos que formam a imagem. Cada pixel é representado por um valor numérico que indica sua intensidade ou cor.

#### Camada de Convolução

A camada de convolução é o bloco de construção central da CNN. Ele carrega a parte principal da carga computacional da rede. Essa camada fica responsável por realizar a convolução entre filtros e a imagem de entrada, permitindo a extração de características visuais relevantes.

A camada convolucional é um conjunto de filtros convolucionais, também chamados de kernels ou janelas de convolução. Cada filtro é uma pequena matriz de pesos que desliza pela imagem de entrada, realizando a operação de convolução em cada local para obter uma resposta.

<img src="https://drive.google.com/uc?id=1LUulDssbcwTKpT5zqGiwMM_PMSxgzAjn" width="400" align="center"/>

Imagem: [Operação de Convolução](https://towardsdatascience.com/convolutional-neural-networks-explained-9cc5188c4939)

A convolução é uma operação matemática que combina os valores dos pixels da imagem com os pesos do filtro em cada local. O resultado é uma nova matriz chamada de feature map ou mapa de características. Cada valor no mapa de características representa a resposta do filtro em uma determinada posição da imagem.

Existem três partes em uma convolução: entrada (por exemplo, imagem 2D), um filtro (também conhecido como kernel) e uma saída (também conhecida como _convolved feature_).

<img src="https://drive.google.com/uc?id=1yu72CNhG_0QMlxfwlgubXuKdqlmikJR1" width="400" align="left"/>

Imagem: [Convolução em funcionamento](https://solclover.com/)

O filtro, também conhecido como kernel ou janela de convolução, é uma pequena matriz de pesos que é aplicada à imagem de entrada por meio da operação de convolução. Cada filtro desliza pela imagem, realizando multiplicações ponto a ponto com os valores dos pixels e produzindo uma resposta para cada posição. Essa resposta é então usada para formar o mapa de características.

A função do filtro é detectar padrões e características específicas nas imagens de entrada. Durante o treinamento da CNN, os valores dos pesos do filtro são ajustados automaticamente através do processo de retropropagação do gradiente, permitindo que a rede neural aprenda a identificar e extrair padrões relevantes para a tarefa em questão.

### Pooling Layer

O Pooling Layer (agrupamento de camadas), também conhecido como _downsampling_, é responsável pela redução de dimensionalidade, reduzindo o número de parâmetros na entrada. É muito comum adicionar uma camada Pooling após uma camada convolucional. Isso ajuda a reduzir o tamanho dos parâmetros convoluídos e melhora a eficiência computacional. É muito útil para extrair características dominantes que são invariantes rotacionais e posicionais, mantendo assim o processo de treinamento efetivo do modelo.

Existem dois tipos principais de agrupamento:

- **Max pooling:** À medida que o filtro se move pela entrada, ele seleciona o pixel com o valor máximo para enviar para a matriz de saída. 

- **Average pooling:** À medida que o filtro se move pela entrada, ele calcula o valor médio dentro do campo receptivo para enviar para a matriz de saída.

Embora muitas informações sejam perdidas na camada de pooling, ela também traz vários benefícios para a CNN. Eles ajudam a reduzir a complexidade, melhorar a eficiência e limitar o risco de overfitting.

### Camadas Totalmente Conectadas (Fully Connected Layer)

Nas camadas totalmente conectadas, cada neurônio recebe como entrada os valores de saída de todos os neurônios da camada anterior. Cada conexão entre os neurônios é associada a um peso, que é ajustado durante o treinamento da rede neural. A função das camadas totalmente conectadas em uma CNN é transformar as características extraídas das camadas anteriores em uma representação adequada para a tarefa de classificação ou regressão. Essas camadas permitem que a rede neural aprenda relações não lineares entre as características extraídas e as classes ou rótulos desejados.

Adicionar uma camada totalmente conectada é uma maneira (geralmente) mais barata de aprender combinações não lineares dos recursos de alto nível, como a saída da camada convolucional. 



### Função de Ativação

As funções de ativação são usadas para introduzir a não-linearidade nas camadas da rede neural, permitindo que a rede aprenda a modelar relacionamentos complexos entre as características das imagens de entrada.

Nas camadas convolucionais de uma rede neural convolucional (CNN), a função de ativação é geralmente aplicada após a operação de convolução. A função de ativação é aplicada elemento a elemento, ou seja, a função é aplicada a cada valor do mapa de características resultante da convolução.

A função de ativação mais usada em camadas convolucionais é a função ReLU (Rectified Linear Unit). A função ReLU define que se o valor de entrada for maior que zero, o valor de saída será igual ao valor de entrada; caso contrário, o valor de saída será zero. A função ReLU é conhecida por ser computacionalmente eficiente e ajudar a mitigar o problema do desaparecimento do gradiente durante o treinamento da rede.

Além disso, nas camadas totalmente conectadas que podem seguir as camadas convolucionais em uma CNN, também é comum usar funções de ativação, como a função ReLU, sigmoidal ou tangente hiperbólica. Essas funções de ativação são aplicadas aos neurônios das camadas totalmente conectadas para introduzir não-linearidade nas operações de combinação de características.

### Detecção de Veículos usando OpenCV
Exemplo adaptado de [Aman Preet Gulati](https://www.analyticsvidhya.com/blog/author/aman_preet_singh_gulati/)

```python
# Carregar Bibliotecas
from PIL import Image
import cv2
import numpy as np
import requests
```

```python
#Carregar uma imagem com veículos da internet
image = Image.open(requests.get('https://static.cdn.pleno.news/2017/09/transito_sp_ibge-1024x682.jpg', stream=True).raw)
image = image.resize((450,250))
image_arr = np.array(image)
image
```

```python
#Converter a imagem para escala de cinza
grey = cv2.cvtColor(image_arr,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)
```

Vamos aplicar uma função camada **GaussianBlur** para remover o ruído da imagem. O desfoque gaussiano é uma das técnicas de processamento de imagem que é amplamente usado em design gráfico para reduzir o ruído e suavizar a imagem, de modo que, para um pré-processamento futuro, gere uma saída melhor. Além de reduzir o ruído na imagem, a técnica de desfoque gaussiano também reduz os detalhes da imagem.

```python
blur = cv2.GaussianBlur(grey,(5,5),0)
Image.fromarray(blur)
```

No próximo passo vamos expandir a imagem. A dilatação é uma das técnicas morfológicas onde tentamos preencher os pixels com o elemento, também conhecidos como kernels (peças estruturadas), para preencher as partes que faltam nas imagens sempre que necessário.

```python
dilated = cv2.dilate(blur,np.ones((3,3)))
Image.fromarray(dilated)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel) 
Image.fromarray(closing)
```

```python
car_cascade_src = 'cars.xml'
car_cascade = cv2.CascadeClassifier(car_cascade_src)
cars = car_cascade.detectMultiScale(closing, 1.1, 1)
cars
```

```python
cnt = 0
for (x,y,w,h) in cars:
    cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
    cnt += 1
print(cnt, " cars found")
Image.fromarray(image_arr)
```

#### Detecção de ônibus

```python
# Ler a imagem da URL
image2 = Image.open(requests.get('https://www.redebrasilatual.com.br/wp-content/uploads/2014/01/59ba15f0-55f9-481e-b23f-e6b72e3e3ad8.jpeg', stream=True).raw)
image2 = image2.resize((450,250))
image_arr2 = np.array(image2)
grey2 = cv2.cvtColor(image_arr2,cv2.COLOR_BGR2GRAY)
```

```python
bus_cascade_src = 'Bus_front.xml'
bus_cascade = cv2.CascadeClassifier(bus_cascade_src)
bus = bus_cascade.detectMultiScale(grey2, 1.1, 1)
```

```python
cnt = 0
for (x,y,w,h) in bus:
  cv2.rectangle(image_arr2,(x,y),(x+w,y+h),(255,0,0),2)
  cnt += 1
print(cnt, " ônibus encontrado")
Image.fromarray(image_arr2)
```

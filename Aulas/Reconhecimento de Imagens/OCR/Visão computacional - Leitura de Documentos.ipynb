#### Exemplo de Detecção de faces usando opencv e Haar Cascade

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Abrir imagem
img = cv2.imread('5ef850b8-f6d3-4e12-a6c0-2a43672abb13.jpg')

# Detectar faces
faces = face_cascade.detectMultiScale(image = img, scaleFactor = 1.1, minNeighbors = 1)

# Desenhar retângulos
for (x, y, w, h) in faces:
      cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Mostrar o número de faces detectadas
print(len(faces),"faces detectadas!")

# Imprimir imagem
finalimg = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.figure(figsize=(5,5))
plt.imshow(finalimg) 
plt.axis("off")
plt.show()
```

### OCR
O OCR (Optical Character Recognition) é um processo que permite a leitura e a conversão de textos  em documentos físicos ou imagens em texto editável. É uma técnica utilizada na visão computacional e na inteligência artificial para extrair informações de documentos e transformá-las em dados estruturados.

O OCR utiliza algoritmos e modelos de aprendizado de máquina para analisar imagens de documentos e identificar os caracteres individuais presentes nelas. O processo envolve etapas como pré-processamento da imagem, segmentação de texto, reconhecimento dos caracteres e pós-processamento dos resultados.

Esse tipo de ferramenta é capaz de reconhecer diferentes tipos de fontes, tamanhos e estilos de caracteres, e também pode lidar com documentos multilíngues. Além disso, os avanços recentes na área incorporam técnicas de processamento de linguagem natural para melhorar a precisão e a compreensão do texto extraído e técnicas avançadas de redes neurais.

```python
import cv2
import numpy as np
import pytesseract
import pandas as pd
import os
```

```python
# Leitura da imagem que será alinhada
imgTest = cv2.imread('preenchido_page-0001.jpg')

# Imagem de referência
imgRef = cv2.imread('template_page-0001.jpg')
```

```python
# Converter imagem para escala de cinza
imgTest_cinza = cv2.cvtColor(imgTest,cv2.COLOR_BGR2GRAY)
imgRef_cinza = cv2.cvtColor(imgRef, cv2.COLOR_BGR2GRAY)
height, width = imgRef_cinza.shape
```

Vamos utilizar um algoritmo chamado SIFT (Scale-Invariant Feature Transform). O SIFT é projetado para identificar pontos de interesse robustos e invariantes a mudanças de escala, rotação, iluminação e outros tipos de transformações geométricas. Ele é amplamente utilizado em aplicações como correspondência de imagens, reconhecimento de objetos e panoramas.

```python
# Configurar o algoritmo SIFT com 1000 características.
sift_detector = cv2.SIFT_create(1000)
```

```python
# Extrair os pontos-chave e descritivos de cada imagem
keyPoint1, des1 = sift_detector.detectAndCompute(imgTest_cinza, None)
keyPoint2, des2 = sift_detector.detectAndCompute(imgRef_cinza, None)
```

```python
# Exibir os ponto chaves de para a imagem de referência na cor verde
#imgKp_Ref = cv2.drawKeypoints(imgRef, keyPoint1, 0, (0, 222, 0), None)
#imgKp_Ref = cv2.resize(imgKp_Ref, (width//2, height//2))

#cv2.imshow('Key Points', imgKp_Ref)
#cv2.waitKey(0)
```

```python
# Combinar as características entre as duas imagens utilizando a função Brute Force Matcher com distância L1
matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
 
# Combinar os dois conjuntos de descritores
matches = matcher.match(des1, des2)

# Ordenar as combinações baseadas na distância Hamming
matches = sorted(matches, key=lambda x: x.distance)

# Pegar as top 90% combinações
matches = matches[:int(len(matches) * 0.9)]
no_of_matches = len(matches)
 
# Mostrar as 20 melhores {good[:20}
#imgMatch = cv2.drawMatches(imgTest, keyPoint2, imgRef, keyPoint1, matches[:100], None, flags=2)
#imgMatch = cv2.resize(imgMatch, (width//3, height//3))
#
#cv2.imshow('Image Match', imgMatch)
#cv2.waitKey(0)
```

```python
# Definir uma matriz vazia 2x2
p1 = np.zeros((no_of_matches, 2))
p2 = np.zeros((no_of_matches, 2))
 
# Armazenar os valores nessas matrizes
for i in range(len(matches)):
    p1[i, :] = keyPoint1[matches[i].queryIdx].pt
    p2[i, :] = keyPoint2[matches[i].trainIdx].pt

#Encontrar a matriz de homografia 
homography, mask = cv2.findHomography(p1, 
                                      p2, 
                                      cv2.RANSAC)
```

A matriz de homografia representa a relação entre pontos correspondentes em duas imagens, levando em consideração a perspectiva e a geometria das cenas. Ela é usada para **mapear pontos de uma imagem para pontos correspondentes em outra imagem**, levando em conta rotação, escala, translação e deformações projetivas. Saiba mais em: [Basic concepts of the homography](https://docs.opencv.org/4.x/d9/dab/tutorial_homography.html)

```python
# Utilizar a matriz de homografia para transformar os dados desalinhados com a imagem de referência
aligned_img = cv2.warpPerspective(imgTest, 
                                  homography, 
                                  (width, height))

# Redimensionar a imagem para combinar com as coordenadas da imagem de correspondência
aligned_img = cv2.resize(aligned_img, (width // 3, 
                                       height // 3))
```

```python
# Converter a imagem para a escala de cinza
aligned_img_grey = cv2.cvtColor(aligned_img, cv2.COLOR_BGR2GRAY)

# Convertendo imagem cinza em imagem binária por Thresholding
aligned_img_threshold = cv2.threshold(aligned_img_grey, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
 
# Copia da imagem de entrada
imgTest_cp = imgTest.copy()
imgTest_cp = cv2.resize(imgTest_cp, (width // 3, height // 3))
# Salvar a imagem alinhada
cv2.imwrite('output.jpg', aligned_img)
```

### Extrair as informações utilizando OCR

```python
# Abrir a imagem com o formulário preenchido
img2 = cv2.imread('preenchido_page-0001.jpg')
cv2.imwrite('preenchido_resized_image.jpg', img2)
```

```python
# Informar as coordenadas de cada elemento a ser mapeado
localizacao_elementos = [
    [(654,45), (786,70), 'text', 'numero'],
    [(619,244), (785,266), 'text', 'data_da_notificacao'],
    [(87,365), (602,387), 'text', 'nome_paciente'],
] 
```

```python
# Carregar o formulário preenchido já redimensionado e corrigida a posição
aligned_img = cv2.imread('preenchido_resized_image.jpg')
aligned_img_show = aligned_img.copy()
aligned_img_mask = np.zeros_like(aligned_img_show)
```

```python
# Informar a localização do executável do Tesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
entity_type = []
entity_name = []
entity_value = []

for roi in localizacao_elementos:
    top_left_x = roi[0][0]
    top_left_y = roi[0][1]
    bottom_right_x = roi[1][0]
    bottom_right_y = roi[1][1]
    # top_left_point = (roi[0][0], roi[0][1])
    # bottom_right_point = (roi[1][0], roi[1][1])
 
    cv2.rectangle(aligned_img_mask, 
                  (top_left_x, top_left_y), 
                  (bottom_right_x, bottom_right_y), 
                  (0, 0, 255), 
                  cv2.FILLED)
    
    aligned_img_show = cv2.addWeighted(aligned_img_show, 
                                       0.99, 
                                       aligned_img_mask, 
                                       0.1, 
                                       0)
 
    # Cortar a parte necessária específica de toda a imagem
    img_cropped = aligned_img_threshold[top_left_y:bottom_right_y, 
                                        top_left_x:bottom_right_x]
 
    # Configuração dos parâmetros do Tesseract
    custom_config = r'--oem 3 --psm 6'
 
    # Extrair texto dentro das coordenadas específicas usando 
    # pytesseract OCR Python Fornecendo imagem recortada como entrada para o OCR
    ocr_output = pytesseract.image_to_string(img_cropped, 
                                             config=custom_config, 
                                             lang='por')
    
    # Remover lacunas de linha extras entre as frases
    cleaned_output = os.linesep.join([s for s in ocr_output.splitlines() if s])
    print([ocr_output])
    print([cleaned_output])
 
    # OpenCV fonte
    font = cv2.FONT_HERSHEY_DUPLEX
    
    # cor vermelha
    red_color = (0, 0, 255)
    
    # Extrair o texto dentro do bloco vermelho
    cv2.putText(aligned_img, f'{cleaned_output}', (top_left_x, top_left_y - 30), font, 0.5, red_color, 1)
 
    # -----------------------------------------------------------
    # Armazenar a saída do OCR em uma lista
    entity_type.append(roi[2])
    entity_name.append(roi[3])
    entity_value.append(cleaned_output)
```

_Nota: Os argumentos `--oem` e `--psm` são parâmetros de linha de comando do Tesseract. 
- `--oem` (Engine Mode) refere-se ao modo de motor do Tesseract e define qual algoritmo de reconhecimento será usado. 
- `--psm` (Page Segmentation Mode) especifica o modo de segmentação de página do Tesseract, que define como o texto é segmentado e organizado em blocos, linhas e palavras._

```python
# -----------------------------------------------------------
# Armazenar o OCR em um DataFrame
form_data = pd.DataFrame()
form_data['Entity_Name'] = entity_name
form_data['Entity_Value'] = entity_value
form_data['Entity_Type'] = entity_type
 
# Save extracted from as csv file
form_data.to_excel('form_data.xlsx', index=False)
 
#cv2.imshow('Input Image', imgTest_cp)
#cv2.imshow('Output Image', aligned_img)
#cv2.imshow('Masked Image', aligned_img_show)
#cv2.waitKey(0)
```
